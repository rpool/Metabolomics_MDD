```python
import numpy
import os
import pandas
import zipfile
```

```python
ZF = zipfile.ZipFile(
    file='./Input/Mtblmcs_Personal20240917_PGCMetaMetabolomics.zip',
    mode='r')

FH = ZF.open(name='DBDF.bsng.tsv',
             mode='r')
MtblmcsVlsDF = pandas.read_csv(filepath_or_buffer=FH,
                               sep='\t')
FH.close()
del FH

FH = ZF.open(name='DBColumnsDF.bsng.tsv',
             mode='r')
MtblmcsClmnsDF = pandas.read_csv(filepath_or_buffer=FH,
                                 sep='\t')
FH.close()
del FH

FH = ZF.open(name='MetaboliteClasses.bsng.tsv',
             mode='r')
MtblmcsClsssDF = pandas.read_csv(filepath_or_buffer=FH,
                                 sep='\t')
FH.close()
del FH

ZF.close()
del ZF
```

```python
Path = os.path.join('/home',
                    'rpool',
                    'ownCloud',
                    'Shared',
                    'PRS_Allelica',
                    'documentation',
                    'DataRequestNTR',
                    'DataDownload',
                    'PHE_20240405_4402_LL.sav')
PhenoDF = pandas.read_spss(path=Path)
PhenoDF['FISNumber'] = PhenoDF['FISNumber'].astype(int)
PhenoDF['lifetime_mdd'].dropna()
PhenoDF['MDD_status'] = numpy.nan
F = PhenoDF['lifetime_mdd'].eq('no')
PhenoDF.loc[F,
            'MDD_status'] = 0.0
del F
F = PhenoDF['lifetime_mdd'].eq('yes')
PhenoDF.loc[F,
            'MDD_status'] = 1.0
del F
```

```python
print(MtblmcsVlsDF.shape)
print(PhenoDF.shape)
MDF = MtblmcsVlsDF.merge(right=PhenoDF[['FISNumber',
                                        'MDD_status']],
                         left_on=['fisnumber'],
                         right_on=['FISNumber'])
print(MDF.shape)
print(MDF.dropna(subset=['MDD_status']).shape)
print(MDF['MDD_status'].value_counts())
```

```python
Columns = pandas.Series(MDF.columns)
Columns.index = Columns.values
Columns = Columns.str.replace(pat='^bsng_',
                              repl='',
                              regex=True)
MDF = MDF.rename(columns=Columns.to_dict())
del Columns
MDF
```

```python
print(MDF.shape)
DropColumns = numpy.intersect1d(ar1 = MtblmcsClsssDF.loc[~MtblmcsClsssDF['boInclude'],
                                                   'Name1'],
                                ar2 = MDF.columns).tolist()
MDF = MDF.drop(labels=DropColumns,
               axis=1)
del DropColumns
print(MDF.shape)
```

```python
Columns = pandas.Series(MDF.columns)
F = Columns.str.contains(pat='indicator')
F = F | Columns.str.contains(pat='umber')
F = F | Columns.str.contains(pat='PID')
F = F | Columns.str.contains(pat='MDD_status')
F.value_counts()
```

```python
MDF.to_csv(path_or_buf='./Output/MDF.tsv',
           sep='\t',
           index=False)
```
